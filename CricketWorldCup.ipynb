## India Win Analysis

```python
India = odi_total[odi_total['Country'] == 'India']
```

```python
India
```

```python
India_win = India[India['Result'] == 'won']
```

```python
India_win
```

```python
India_win['Opposition'].value_counts()
```

```python
plt.figure(figsize = (8, 5))
plt.bar(list(India_win['Opposition'].value_counts()[0:5].keys()), list(India_win['Opposition'].value_counts()[0:5]), color = 'g')
plt.show()
```

```python
India_lost = India[India['Result'] == 'lost']
```

```python
India_lost['Opposition'].value_counts()
```

```python
plt.figure(figsize = (8, 5))
plt.bar(list(India_lost['Opposition'].value_counts()[0:5].keys()), list(India_lost['Opposition'].value_counts()[0:5]))
plt.show()
```

```python
# making a pie chart on loss of India
plt.figure(figsize = (7, 7))
plt.pie(list(India_lost['Opposition'].value_counts()), labels = list(India_lost['Opposition'].value_counts().keys()), autopct = '%0.1f%%')
plt.title('Losses of India in WorldCup')
plt.show()
```

```python
# making a pie chart on India won
plt.figure(figsize = (7, 7))
plt.pie(list(India_win['Opposition'].value_counts()), labels = list(India_win['Opposition'].value_counts().keys()), autopct = '%0.1f%%')
plt.title('Winning Streek of India')
plt.show()
```
